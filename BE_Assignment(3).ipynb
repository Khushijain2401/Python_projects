```python
import matplotlib.pyplot as plt
```

```python
# Years vs Temperature
years = [2010, 2011, 2012, 2013, 2014]
temperature = [22, 23, 21, 24, 25]
plt.plot(years,temperature)

plt.xlabel('Year')
plt.ylabel('Temperature(Â°C)')
plt.title("Average Temperature Over Years")

plt.show()
```

```python
# Height vs Weight
height = [150, 160, 165, 170, 175, 180]
weight = [50, 55, 60, 65, 70, 75]
plt.scatter(height,weight)
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.title("Height vs Weight")
plt.show()
```

```python
# Subjects and student count
subjects = ['Math', 'Science', 'English', 'History']
students = [30, 25, 35, 20]
plt.bar(subjects,students)
plt.xlabel("Subjects")
plt.ylabel("Students")
plt.title("Number of Students per Subject")
plt.show()
```

```python
# Activities and time
activities = ['Sleep', 'Study', 'Exercise', 'Entertainment']
hours = [8, 6, 2, 8]

plt.pie(hours,labels=activities, autopct="%1.1f%%")
plt.title("Daily Activity Distribution")
plt.show()
```
